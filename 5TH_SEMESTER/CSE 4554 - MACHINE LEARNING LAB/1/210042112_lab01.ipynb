**Task 1-Get all the items between 5 and 10 from a. a = np.array([2, 6, 1, 9, 10, 3, 27])**

```python
a=np.array([2, 6, 1, 9, 10, 3, 27])
result= a[(a >= 5) & (a <= 10)] #(a>=5) returns the numbers true for the fitting condition and (a<=10) does the same.They are boolean expressions.
print(result)
```

**Task 6- Find the mean, median and standard deviation of the 1st column of the given dataset.
 Give statistical summary of the dataset as many as possible.**

```python
import numpy as np
from sklearn import datasets

iris = datasets.load_iris()

# Extract numerical data
data = iris.data

# First column (sepal.length)
sepal_length = data[:, 0]  #all rows and first column

mean_sepal_length = np.mean(sepal_length)
median_sepal_length = np.median(sepal_length)
std_sepal_length = np.std(sepal_length)

print(f"Sepal Length - Mean: {mean_sepal_length:.6f}, Median: {median_sepal_length:.6f}, Std Dev: {std_sepal_length:.6f}")

# statistical summary
mean_all = np.mean(data, axis=0) #every column

median_all = np.median(data, axis=0)

std_all = np.std(data, axis=0)

variance_all = np.var(data, axis=0)

min_all = np.min(data, axis=0)

max_all = np.max(data, axis=0)

range_all = max_all - min_all


print("\nStatistical Summary of the Dataset (All Columns):\n")
print(f"Mean: {mean_all}")
print(f"Median: {median_all}")
print(f"Standard Deviation: {std_all}")
print(f"Variance: {variance_all}")
print(f"Minimum: {min_all}")
print(f"Maximum: {max_all}")
print(f"Range: {range_all}")
```

**Task 3 : Suppose you are an investor and want to invest in housing properties. Use this dataset
 to make decision on the location of investing.**

```python
 import matplotlib.pyplot as plt

 plt.figure(figsize=(10, 6))
 plt.scatter(df2['median_income'], df2['median_house_value'], alpha=0.5, c=df2['population'], cmap='inferno')
 plt.colorbar(label='Population')
 plt.title('Median Income vs Median House Value')
 plt.xlabel('Median Income (in tens of thousands)')
 plt.ylabel('Median House Value (in dollars)')
 plt.grid(True)
 plt.show()
```

* I think the best option for investing would be to focus on areas with a median income between 6 to 10 on the x-axis and a median house value ranging from 200000 to 400000.

* These areas are likely to have people with stable and decent earning capacity, which supports steady demand for housing. It's a more favourable spot where properties are affordable for many buyers or renters but still likely to appreciate over time.

* Properties in the 200000-400000 range are not hitting the apparent 500000 value cap, which could mean they likely have room for appreciation in value. These mid-range houses may attract more buyers than premium properties, especially in growing markets.

* These areas have moderate population. These areas may offer a balance between available housing and demand, allowing an easier entry into the market and generate good returns in growing markets.
